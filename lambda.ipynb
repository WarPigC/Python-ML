```python
l = list(range(1, 10))
list(filter(lambda x : x & 1 == 0, l))
```

```python
list(filter(lambda x: x > 5, l))
```

```python
class test:
  def __init__(self):
    self.var = 5

  def getValue(self):
    print(self.var)

test().getValue()
```

```python
def addition(n):
  return n * 2

num = tuple(range(1, 5))
print(list(map(addition, num)))
```

```python
print(
    list(map(list, ['1234', 'bat', 'cat', 'mat']))
)
```

```python
cube = lambda y: y ** 3

print(cube(5))
```

```python
# def func(x): return x * 10

# tables = [func(y) for y in range(1, 10 + 1)]

tables = [lambda x = y  : x * 10 for y in range(1, 10 + 1)]
for i in tables:
  print(i())
```

```python
l = [[4, 3, 2], [5, 4, 3, 2, 1]]
Sorted = lambda x: [sorted(i) for i in x]
print(Sorted(l))
```

```python
from functools import reduce
li = [5, 6, 7, 8, 9]
sum = reduce(lambda x, y: x + y, li)
print(sum)
```
